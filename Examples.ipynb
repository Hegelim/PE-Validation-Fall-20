## Example of checklist

Example file: Covid_Reasoningv1-2020-10-04T1803-DataHunt.csv <br>
Question: 1 (cl)

```python
reasoning = pd.read_csv("Specialists data/Covid_Reasoningv1-2020-10-04T1803-DataHunt.csv")
reasoning = reasoning[reasoning["question_label"] == "T1.Q1"]
reasoning = reasoning.reindex(columns=cols)
# reasoning.to_excel("example1.xlsx")
example = pd.get_dummies(reasoning, columns = ["answer_label"])
# example.to_excel("example2.xlsx")
```

```python
example_three = example.groupby(["quiz_task_uuid", "contributor_uuid"]).agg(sum)
example_three = example_three.reset_index()
# example_three.to_excel("example_three.xlsx")
```

```python
example_four = pd.melt(example_three, id_vars=["quiz_task_uuid", "contributor_uuid"], 
                                  var_name = "choice", 
                                  value_name = "value")
example_four = example_four.sort_values(by=["quiz_task_uuid", "contributor_uuid", "choice"])
# example_four.to_excel("example_four.xlsx")
```

```python
example_five = example_four.copy()
example_five["choice"] = example_five["quiz_task_uuid"].str.slice(start=0, stop=6) + " " + example_five["choice"]
example_five.head()
```

```python
example_six = pd.pivot_table(example_five, values='value', index='contributor_uuid', 
                          columns='choice', 
                          aggfunc=select)
example_six.to_excel("example_6.xlsx")
```

```python
example_six.head()
```

```python
array = example_six.to_numpy()
```

```python
array
```

```python
print(round(alpha(reliability_data=array, level_of_measurement='nominal'), 6))
```
